#image that will be generated

```python
#stroing the image in a varible img_response and using function openai.Image.create()
img_response=openai.Image.create(
    prompt='a cat drinking milk', # it is a variable the stores the description of image to be generated one can change it.
    n=1, # how many images we wanna generate (let say 1 at a time)
    size="512x512") # size of the image we want it can be in three formats "256x256" "512x512" "1024x1024"
```

```python
img_response #this is an openai object by above the image has been generated
```

```python
img_response['data'][0]['url'] #it basically fetches the url generated from a list (and within a list a dictionary)
```

```python
img_url = img_response['data'][0]['url'] # storing the url in img_url variable and printing this url
img_url
```

```python
urllib.request.urlretrieve(img_url, "image.png") # it retrives the image and stores in the parameter image.png (image will be created in the folder and saved automatically)
```

```python
img=cv2.imread("image.png") # reading the image from cv2 open library
cv2_imshow(img) # displaying the image
```
