```python
#Zooming
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the original grayscale image
image_path = '/content/c.jpeg'  # Replace with the path to your image
original_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Check if the image was loaded successfully
if original_image is None:
    print("Error: Unable to load the image.")
else:
    # Define the region to zoom in
    x, y, width, height = 00, 00, 800, 900  # Change these values as needed

    # Extract the region of interest (ROI) from the original image
    zoomed_region = original_image[y:y+height, x:x+width]

    # Define the zoom factor
    zoom_factor = 2  # Adjust as needed

    # Create a blank canvas for the zoomed image
    zoomed_image = np.zeros((height * zoom_factor, width * zoom_factor), dtype=np.uint8)

    # Perform manual zooming without inbuilt functions
    for i in range(zoomed_image.shape[0]):
        for j in range(zoomed_image.shape[1]):
            x_original = int(j / zoom_factor) + x
            y_original = int(i / zoom_factor) + y
            if 0 <= x_original < original_image.shape[1] and 0 <= y_original < original_image.shape[0]:
                zoomed_image[i, j] = original_image[y_original, x_original]

    # Create a figure with two subplots to display images side by side
    fig, axes = plt.subplots(1, 2, figsize=(12, 6))

    # Display the original grayscale image on the left subplot
    axes[0].imshow(original_image, cmap='gray')
    axes[0].set_title('Original Image')

    # Display the manually zoomed grayscale image on the right subplot
    axes[1].imshow(zoomed_image, cmap='gray')
    axes[1].set_title('Zoomed Image')

    # Remove axis labels and adjust spacing
    for ax in axes:
        ax.axis('off')

    # Adjust subplot spacing
    plt.subplots_adjust(wspace=0.05)

    # Show the plot
    plt.show()
```
