```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Load the input image and preprocess it
im = cv2.imread('/content/WhatsApp Image 2023-08-21 at 1.11.01 AM.jpeg')
im = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
im = cv2.resize(im, (100, 100))

# Display the original grayscale image
plt.figure()
plt.imshow(im, cmap='gray')
plt.title('Original Grayscale Image')
plt.show()

# Define different kernels for various effects
# Gaussian Blur kernel
gaussian_kernel = np.array([
    [1/16, 1/8, 1/16],
    [1/8 , 1/4, 1/8],
    [1/16, 1/8, 1/16]
])

# Vertical Edge kernel
vertical_edge_kernel = np.array([
    [1, 0, -1],
    [1, 0, -1],
    [1, 0, -1]
])

# Outline kernel
outline_kernel = np.array([
    [-1, -1, -1],
    [-1,  8, -1],
    [-1, -1, -1]
])

# Initialize an array for storing convolution results
convolved_image = np.zeros((im.shape[0] - outline_kernel.shape[0] + 1, im.shape[1] - outline_kernel.shape[1] + 1))

# Perform convolution using the Outline kernel
for i in range(convolved_image.shape[0]):
    for j in range(convolved_image.shape[1]):
        patch = im[i:i+outline_kernel.shape[0], j:j+outline_kernel.shape[1]]
        hprod = (patch * outline_kernel).sum()
        convolved_image[i, j] = hprod

# Display the convolved image using the Outline kernel
plt.figure()
plt.imshow(convolved_image, cmap='gray')
plt.title('Convolved Image with Outline Kernel')
plt.show()
```

```python
#extending it for vertical edge kernel and gaussian blur

# Perform convolution using the Vertical Edge kernel
convolved_image_vertical_edge = np.zeros((im.shape[0] - vertical_edge_kernel.shape[0] + 1, im.shape[1] - vertical_edge_kernel.shape[1] + 1))

for i in range(convolved_image_vertical_edge.shape[0]):
    for j in range(convolved_image_vertical_edge.shape[1]):
        patch = im[i:i+vertical_edge_kernel.shape[0], j:j+vertical_edge_kernel.shape[1]]
        hprod = (patch * vertical_edge_kernel).sum()
        convolved_image_vertical_edge[i, j] = hprod

# Performing  convolution using the Gaussian Blur kernel
convolved_image_gaussian = np.zeros((im.shape[0] - gaussian_kernel.shape[0] + 1, im.shape[1] - gaussian_kernel.shape[1] + 1))

for i in range(convolved_image_gaussian.shape[0]):
    for j in range(convolved_image_gaussian.shape[1]):
        patch = im[i:i+gaussian_kernel.shape[0], j:j+gaussian_kernel.shape[1]]
        hprod = (patch * gaussian_kernel).sum()
        convolved_image_gaussian[i, j] = hprod

# Displaying the convolved images using different kernels
plt.figure(figsize=(12, 4))

plt.subplot(131)
plt.imshow(convolved_image_vertical_edge, cmap='gray')
plt.title('Convolved Image with Vertical Edge Kernel')
plt.subplot(133)
plt.imshow(convolved_image_gaussian, cmap='gray')
plt.title('Convolved Image with Gaussian Blur Kernel')

plt.tight_layout()
plt.show()
```
